```python
df_clean['Courier Status'] = df_clean['Courier Status'].fillna('Not Assigned')
```

```python
df_clean.info()
```

```python
df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)
```

```python
df_clean = df_clean.drop(columns=['index'])
df_clean.reset_index(drop=True, inplace=True)
```

```python
state_mapping = {
    "Orissa":"Odisha",
    "Rajsthan":"Rajasthan",
    "Rajshthan":"Rajasthan",
    "Rj":"Rajasthan",
    "Pb":"Punjab",
    "Pondicherry":"Puducherry"
}

df_clean["ship_state"] = df_clean["ship_state"].replace(state_mapping)
```
